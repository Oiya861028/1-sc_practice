```python
import pandas as pd
from pathlib import Path
from load_pseudobulk_filters import load_filters
```

```python
df_list_pre, df_list_post = load_filters(merge=False)
```

```python
len(df_list_post)
```

```python
lfc_threshold = 0.5
padj_threshold = 0.05
for i, df in enumerate(df_list_pre):
    # A column of same size as row, containing True or False depending on what is considered differentially expressed
    differential_expressed_mask = (df["padj"] < padj_threshold) & (abs(df["log2FoldChange"]) > lfc_threshold)
    df["Significance"] = differential_expressed_mask

    # Adding the column that express differential expressing using words. Ie, up for upregulated, down for downregulated, NS for not significant
    df["diff_expr"] = "NS"
    df["diff_expr"] = df["diff_expr"].case_when([
        (differential_expressed_mask & (df["log2FoldChange"] > lfc_threshold), "up"),
        (differential_expressed_mask & (df["log2FoldChange"] < -lfc_threshold), "down"),
    ])

    # Saving the updated file in a temperary location for double check
    df.to_csv("Prefilter/DEG_Hindbrain_Control_vs_SmoNull_"+str(i)+"_cells")


for i, df in enumerate(df_list_post):
    # A column of same size as row, containing True or False depending on what is considered differentially expressed
    differential_expressed_mask = (df["padj"] < padj_threshold) & (abs(df["log2FoldChange"]) > lfc_threshold)
    df["Significance"] = differential_expressed_mask

    df["diff_expr"] = "NS"
    df["diff_expr"] = df["diff_expr"].case_when([
        (differential_expressed_mask & (df["log2FoldChange"] > lfc_threshold), "up"),
        (differential_expressed_mask & (df["log2FoldChange"] < -lfc_threshold), "down"),
    ])
    # Saving the updated file in a temperary location for double check
    df.to_csv("Postfilter/DEG_Hindbrain_Control_vs_SmoNull_"+str(i)+"_cells")
```
